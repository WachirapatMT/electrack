```python
import pandas as pd
fname = "HomeC.csv"

df = pd.read_csv(fname)
df.head()
```

```python
n_record = len(df)
n_record
```

```python
import datetime

timestamp = 1451624400
dt_object = datetime.datetime.fromtimestamp(timestamp)
starter_time = dt_object
one_minute = datetime.timedelta(minutes=1)
one_minute * 2
```

```python
col_time = [starter_time + one_minute*n for n in range(n_record)]
col_time = [ x.isoformat() for x in col_time]
col_time
```

```python
df[:20]
```

```python
df['time'] = col_time
df['time']
```

```python
df.head()
```

```python
cols = [col for col in df.columns]
cols = cols[4:]
cols
```

```python
from tqdm import tqdm
```

```python
for col in tqdm(cols):
    t_df = df.filter(items=['time', col])
    t_df['source'] = [col]*n_record
    t_df.to_csv("split_files/" + col + '.csv', index=False, header=False)
```
